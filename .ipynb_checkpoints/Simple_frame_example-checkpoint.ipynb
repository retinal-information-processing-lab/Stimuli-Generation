```python
import numpy as np
from matplotlib import pyplot as plt
import os
```

# Numpy stimulus frame

```python
from npy_stim_gen import *
```

```python
# MEA 2
x_pix = 1920
y_pix = 1080
fdim = 800

# # MEA 1&3
# x_pix = 1024
# y_pix = 768
# fdim = 800

F_frame = create_f(x_pix, y_pix, fdim, thick=5)

plt.imshow(F_frame)
plt.show()
```

# Binary frame

```python
from bin_manipulation import BinFile
```

```python
# Create the bin
bin_path = './F.bin'
frame_xsize, frame_ysize = F_frame.shape

F_file = BinFile(bin_path, frame_xsize=frame_xsize, frame_ysize=frame_ysize, nb_images=1, reverse=False, mode='w')

# Append the frame
F_file.append(F_frame.astype('uint8'))
F_file.close()
```

```python
# Reload the bin from the written file
F_file = BinFile(bin_path, frame_xsize=frame_xsize, frame_ysize=frame_ysize, reverse=False, mode='r')
# Read and plot the frame
F_read = F_file.read_frame(0)

plt.figure(figsize=(6,5), dpi=100)

plt.subplot(121)
plt.title('Npy stimulus frame')
plt.imshow(F_frame)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(122)
plt.title('Reloaded from .bin frame')
plt.imshow(F_read)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.tight_layout()
```
