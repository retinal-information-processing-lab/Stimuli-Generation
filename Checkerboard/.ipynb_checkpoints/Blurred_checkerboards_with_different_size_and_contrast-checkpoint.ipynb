```python
def regular_checkerboard(w, h, c0, c1, blocksize):
        tile = np.array([[c0,c1],[c1,c0]]).repeat(blocksize, axis=0).repeat(blocksize, axis=1)
        grid = np.tile(tile,(int(math.ceil((h+0.0)/(2*blocksize))),int(math.ceil((w+0.0)/(2*blocksize)))))
        return grid[:h,:w]
```

## Generate a single checkerboard from the binary source file, for different check sizes

We want to probe checks with a size around 25, 50 and 100 um. Final checkerboard size will be 864x864 pixels.

```python
nb_checks_by_side[-1]
```

```python
# We generate checkerboards with the right number of checks, but 1 pixel/check,
# and then we will upscale it to get the right number of pixels/check

checkerboard_size = 864
nb_pixels_per_check = np.array([6, 16, 32]) # 21, 56, 112 um
nb_checks_by_side = checkerboard_size/nb_pixels_per_check

checkerboards = []

for i in range(len(nb_checks_by_side)):
    checkerboard = regular_checkerboard(int(nb_checks_by_side[i]), int(nb_checks_by_side[i]), 0, 1, 1)
    checkerboards.append(checkerboard)
```

```python
plt.figure(figsize = (6,6))
plt.imshow(checkerboards[0], cmap = 'Greys')
```

## Generate checkerboards with different contrasts

```python
contrasts = np.linspace(0,1,6)

contrast_checkerboards = []

for ich in range(len(checkerboards)):
    contrast_checkerboard = np.zeros((len(contrasts), len(contrasts), \
                                                       int(nb_checks_by_side[ich]), int(nb_checks_by_side[ich])))
    for i in range(len(contrasts)):
        # Changing the contrast of even pixels
        checkbd = checkerboards[ich] * contrasts[i]
        plt.figure(figsize = (16,5))
        for j in range(len(contrasts)):
            ch = checkbd 
            ch[::2, ::2] = np.flip(contrasts)[j]
            ch[1::2, 1::2] = np.flip(contrasts)[j]
            contrast_checkerboard[i,j,:,:] = ch

            plt.subplot(1,6,j+1)
            plt.imshow(ch, cmap = 'Greys', vmin=0, vmax=1)
            plt.axis('off')
    contrast_checkerboards.append(contrast_checkerboard)
```

## Upscale the checkerboards to the right pixel size

```python
sharp_checkerboards = [] #np.zeros((len(checkerboards), len(contrasts), len(contrasts), \
#                                                        checkerboard_size, checkerboard_size))
sharp_checkerboards_names = []

for ich in range(len(checkerboards)):
    for i in range(len(contrasts)):
        for j in range(len(contrasts)):
            # Rescale the checkerboard to the right pixel size
            im = contrast_checkerboards[ich][i,j,:,:]
            rescaled_image = rescale(im, (nb_pixels_per_check[ich],nb_pixels_per_check[ich]), order = 0)
            #images[ich,i,j,:,:] = rescaled_image
            sharp_checkerboards.append(rescaled_image)
            # Record the name of the checkerboard
            name = 'checkerboard_' + str(nb_pixels_per_check[ich]) + 'pix_' + \
                                str(round(contrasts[i],2)) + '_' + str(round(np.flip(contrasts)[j],2))
            sharp_checkerboards_names.append(name)
            # Save checkerboards
            
            imsave('D:/MEA data/stimulations/checkerboards_20220701/sharp checkerboards/' + name + '.png', \
                                 (rescaled_image * 255).astype('uint8'))

sharp_checkerboards_names = np.array(sharp_checkerboards_names)
sharp_checkerboards = np.array(sharp_checkerboards)
```
